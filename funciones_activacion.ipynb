```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    a=1/(1+np.exp(-x))
    return a
    
```

```python
x=np.linspace(-10,10,100)
```

```python
x
```

```python
plt.plot(x,sigmoid(x))
```

```python
def step(x):
    return np.piecewise(x, [x<0.0,x>0.0],[0,1])
```

```python
plt.plot(x,step(x))
```

```python
def tangents(x):
    return np.tanh(x)

```

```python
plt.plot(x,tangents(x))
```

```python
def mse(y,y_hat,derivada=False):
    if derivada:
        return (y-y_hat)
    else:
        return np.mean((y_hat-y)**2)

```

```python
real=np.array([0,0,1,1])
pred=np.array([0.9,0.5,0.2,0.0])

mse_real=mse(real,pred)
```
